```python
# --- 1. GOOGLE COLAB SETUP (Must run first) ---


"""
PJ Checklist before sending
- Excel distribution_List


Caution
1.don't reduce the time.sleep value , doing so can block access to Gmail
2.Check that the email is sending to the proper address, as the distribution list updates slowly
"""



from google.colab import drive
import sys
import os
import time
from pathlib import Path
# Mount Google Drive
drive.mount('/content/drive')

# --- 2. PATH CONFIGURATION ---
# Root folder: The parent folder containing both 'Mail_sender' and 'PJ_mail'
root_path = Path('/content/drive/MyDrive/Mail_automation_program')

# Working folder: The specific folder where your images, excel, and script logic reside
working_path = Path('/content/drive/MyDrive/Mail_automation_program/PJ_mail')

# A. Allow imports from the root (enables 'from Mail_sender import...')
if root_path not in sys.path:
    sys.path.append(root_path)

# B. Change the "Current Working Directory" to the project folder
# This makes the code find 'distribution_list.xlsx' and images automatically
os.chdir(working_path)

print(f"Current System Path includes: {root_path}")
print(f"Current Working Directory is: {os.getcwd()}")
print("-" * 30)


# --- 3. YOUR ORIGINAL LOGIC (Imports now work) ---

# Now you can import from the sibling folder because 'root_path' is in sys.path
from Mail_sender import gmail_automation

# You can import this because we are inside 'working_path' (os.chdir)
import PJ_mail_writer

# --- Configuration: File Paths (No changes needed here) ---
DISTRIBUTION_LIST_PATH = root_path / "distribution_list.xlsx"
HTML_TEMPLATE_PATH = root_path/"Mail_form"/"PJ_template.html"
PDF_ATTACHMENT_PATH = root_path/"Mail_form"/"서울대학교 전략컨설팅학회 BICS 소개서.pdf"
IMAGE_1_PATH = root_path/"Mail_form"/"bics_intro_1.png"
IMAGE_2_PATH = root_path/"Mail_form"/"bics_intro_2.png"


def main():
    print("--- Starting BICS Email Automation ---")

    # 1. Load Email Configuration
    try:
        config = gmail_automation.load_email_config()
        sender_email = config["sender_email"]
        print(f"Logged in as: {sender_email}")
    except Exception as e:
        print(f"Error loading email configuration: {e}")
        # Tip: Ensure googleid.env is accessible. Since it is in Mail_sender,
        # your gmail_automation.py might need an absolute path adjustment inside itself
        # if it uses relative paths.
        return

    # 2. Load Distribution List
    try:
        recipients = PJ_mail_writer.load_distribution_list(DISTRIBUTION_LIST_PATH)
        print(f"Found {len(recipients)} recipients in {DISTRIBUTION_LIST_PATH}")
    except Exception as e:
        print(f"Error reading distribution list: {e}")
        return

    # 3. Iterate and Send Emails
    success_count = 0
    fail_count = 0

    for i, recipient in enumerate(recipients, 1):
        company = recipient['company']
        email_addr = recipient['email']
        contents = recipient['contents']

        print(f"\n[{i}/{len(recipients)}] Preparing email for {company} ({email_addr})...")

        try:
            # A. Create the full email object
            email_message = PJ_mail_writer.create_project_email(
                sender_email=sender_email,
                recipient_email=email_addr,
                company_name=company,
                custom_contents=contents,
                html_template_path=HTML_TEMPLATE_PATH,
                pdf_path=PDF_ATTACHMENT_PATH,
                img1_path=IMAGE_1_PATH,
                img2_path=IMAGE_2_PATH
            )

            # B. Send the email
            gmail_automation.send_gmail(email_message)
            success_count += 1

            # C. Rate Limiting
            time.sleep(25) #If it is too short, gmail blcok the email

        except Exception as e:
            print(f"Failed to send to {company}: {e}")
            fail_count += 1

    # 4. Final Summary
    print("\n--- Batch Processing Complete ---")
    print(f"Success: {success_count}")
    print(f"Failed:  {fail_count}")

# Run the main function
if __name__ == "__main__":
    main()
time.sleep(20)
```
